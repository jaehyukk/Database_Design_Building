### 1) DB connection check

```python
SERVER_IP = "202.30.21.40"
DB_NAME = "db201821094"
```

```python
import logging
import pymysql

import socket

def _CheckIP():
    return socket.gethostbyname(socket.gethostname()) == SERVER_IP

def _getHostName():    
    if _CheckIP():
        return "localhost"
    else:
        return SERVER_IP
        
def connectDb(user, password):
    host = _getHostName()

    try:
        return pymysql.connect(host = host, user = user, password = password, db = DB_NAME, charset="utf8")
    except:
        logging.error("Check DB info")
        logging.error("user : " + user + ", password : " + password)        
```

```python
conn = connectDb("std201821094", "201821094")
cur = conn.cursor()
```

위의 코드는 HeidiSQL에 접속하기 위한 것으로, 서버 아이디와 비밀번호를 이용하여 연결을 시도한다.

```python
#timewindow table
#첫 200개는 99개씩, 이후로 98개씩 데이터 슬라이싱
import numpy as np

length = 30450 - 850
num_min = 850
num_max = 948
num = np.array([num_min, num_max])
num_list = []

for i in range(0, 300):
    num[0] = num_min
    num[1] = num_max
    print(num)
    num_list.append((num_min, num_max))
    num_min = num_max + 1
    if i < 200:   # 99개씩 슬라이싱 하는 구간
        num_max += 99
    else:         # 98개씩 슬라이싱 하는 구간
        num_max += 98
```

```python
query = "insert into timewindow (start_dataIdx, end_dataIdx) values (%s, %s)"

cur.executemany(query, num_list)
conn.commit()
```

```python
for reg in reg_Name:
    if reg is reg_Name[0]:
        reg_ID = '1'
        query_cal = "SELECT AVG(PORT4_CHN1), SUM(PORT4_CHN1), COUNT(PORT4_CHN1) FROM accelerator WHERE data_idx BETWEEN %s AND %s"
    elif reg is reg_Name[1]:
        reg_ID = '2'
        query_cal = "SELECT AVG(PORT6_CHN1), SUM(PORT6_CHN1), COUNT(PORT6_CHN1) FROM accelerator WHERE data_idx BETWEEN %s AND %s"
    elif reg is reg_Name[2]:
        reg_ID = '3'
        query_cal = "SELECT AVG(PORT7_CHN1), SUM(PORT7_CHN1), COUNT(PORT7_CHN1) FROM accelerator WHERE data_idx BETWEEN %s AND %s"
    
    idx = 1   # window_idx에 값을 넣어주기 위한 변수
    for i in window:
        query =  "insert into statistic (sensor_ID, reg_ID, window_idx, avg, sum, count) values ('1', %s, %s, %s, %s, %s)"
        cur.execute(query_cal, (i[0], i[1]))
        result = cur.fetchall()
        print(result)
        cur.execute(query, (reg_ID, idx, result[0][0], result[0][1], result[0][2]))
        idx += 1
    conn.commit() 
```

### 4) Search table

After defining five arbitrary conditions, search for tuples that satisfy the condition in the statistic table.

A) Condition 1

- 각 레지스터의 전체 평균에 대해 10% 이내의 오차를 가지고 있는 윈도우 개수

```python
import pandas as pd

query = "SELECT AVG(PORT4_CHN1), AVG(PORT6_CHN1), AVG(PORT7_CHN1) FROM accelerator"
cur.execute(query)
AVG = cur.fetchall()
print(pd.DataFrame(AVG, columns = ['PORT4_avg', 'PORT6_avg', 'PORT7_avg']))

for i in AVG[0]:
    if i is AVG[0][0]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='1' AND (AVG < %s*0.90 AND AVG > %s*1.1)"
        print("PORT4_CHN1:")
    elif i is AVG[0][1]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='2' AND (AVG > %s*0.90 AND AVG < %s*1.1)"
        print("PORT6_CHN1:")
    elif i is AVG[0][2]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='3' AND (AVG < %s*0.90 AND AVG > %s*1.1)"
        print("PORT7_CHN1:")

    cur.execute(query2, (i, i))
    result = cur.fetchall()
    print(result[0])
```

=> 첫 번째 조건으로는 각 레지스터의 평균에 대해 10% 이하의 오차를 가지고 있는 경우를 나타내었다. 이는 각 윈도우에 대한 레지스터의 평균은 신뢰구간 90%에 속하는지에 대해 나타낸 것이며, 먼저 각 레지스터에 대해 분류를 해보았다. 결과로 'PORT4_CHN1'와 'PORT7_CHN1'은 신뢰구간 90%에 해당하는 윈도우가 없는 것으로 나타났다. 반대로 'PORT6_CHN1'은 90%에 속하는 윈도우의 개수가 19개이며, 전체적으로 모든 레지스터에 대하여 90%의 신뢰구간에 속하는 레지스터가 매우 적다는 것을 알 수 있다. 이는 전체 데이터에서 윈도우를 나눌 때 첫 인덱스부터 순차적으로 나눈 윈도우로부터 나타난 결과이고, 각 윈도우 구간마다 평균값 차이가 큰 것이 다음과 같은 결과의 원인이라 생각한다.

B) Condition 2

- 각 레지스터의 전체 평균에 대해 5% 이내의 오차를 가지고 있는 윈도우 개수

```python
import pandas as pd

query = "SELECT AVG(PORT4_CHN1), AVG(PORT6_CHN1), AVG(PORT7_CHN1) FROM accelerator"
cur.execute(query)
AVG = cur.fetchall()
print(pd.DataFrame(AVG, columns= ['PORT4_avg', 'PORT6_avg', 'PORT7_avg']))

for i in AVG[0]:
    if i is AVG[0][0]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='1' AND (AVG < %s*0.95 AND AVG > %s*1.05)"
        print("PORT4_CHN1:")
    elif i is AVG[0][1]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='2' AND (AVG > %s*0.95 AND AVG < %s*1.05)"
        print("PORT6_CHN1:")
    elif i is AVG[0][2]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='3' AND (AVG < %s*0.95 AND AVG > %s*1.05)"
        print("PORT7_CHN1:")

    cur.execute(query2, (i, i))
    result = cur.fetchall()
    print(result[0])
```

=> 두 번째 조건으로는 첫 번째 조건에 이어서 각 레지스터의 평균에 대해 5% 이하의 오차를 가지고 있는 경우를 나타내었다. 이는 각 윈도우에 대한 레지스터의 평균이 신뢰구간 95%에 속하는지에 대해 나타낸 것이다. 'PORT4_CHN1'와 'PORT_CHN1'은 신뢰구간 95%에 해당하는 윈도우가 없는 것으로 나타났는데, 평균 오차 10%이내인 경우에서 해당하는 윈도우가 없었으므로 당연한 결과라고 할 수 있다. 'PORT6_CHN1'은 95%에 속하는 윈도우의 개수가 14개로 줄어들었다. 즉 PORT6_CHN1에 대하여 신뢰구간 90%와 95% 사이에서 존재하는 데이터가 총 5개인 것을 알 수 있다. condition 1과 2를 통해, 만일 일정 구간마다 데이터의 값 변화가 크다면, 데이터를 첫 구간부터 마지막 구간까지 일정한 개수로 순차적으로 나누는 과정이 데이터를 정확하게 파악하기에 결코 좋은 방법이 아니라는 것을 알 수 있다. 특히 데이터 수치의 변동이 큰 경우에는 각 윈도우 별로 구한 평균값들이 크게 차이가 나며, 이는 일정한 구간의 데이터가 전체의 데이터 특성을 나타내는 것으로 정의하기에 어려움을 의미한다.

C) Condition 3

- 각 레지스터에 대한 최종 합계보다 작은 값을 가지고 있는 윈도우 개수

```python
import pandas as pd

query = "SELECT SUM(PORT4_CHN1), SUM(PORT6_CHN1), SUM(PORT7_CHN1) FROM accelerator"
cur.execute(query)
SUM = cur.fetchall()
print(pd.DataFrame(SUM, columns= ['PORT4_sum', 'PORT6_sum', 'PORT7_sum']))

for i in SUM[0]:
    if i is SUM[0][0]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='1' AND SUM < %s"
        print("PORT4_CHN1:")
    elif i is SUM[0][1]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='2' AND SUM < %s"
        print("PORT6_CHN1:")
    elif i is SUM[0][2]:
        query2 = "SELECT COUNT(AVG) FROM statistic where sensor_ID='1' AND reg_ID='3' AND SUM < %s"
        print("PORT7_CHN1:")

    cur.execute(query2, i)
    result = cur.fetchall()
    print(result[0]) 
```

=> 세 번째 조건으로는, 각 레지스터에 대해 전체 데이터의 합계보다 작은 합계를 가지고 있는 윈도우의 개수를 파악하는 과정이다. 이는 모든 레지스터에 대해 앞서 신뢰구간 90%와 95%에 해당하는 윈도우가 매우 적은 이유를 알고자 진행한 과정이라 할 수 있다. 먼저 PORT4_CHN1, PORT6_CHN1, PORT7_CHN1에 대한 전체 데이터 합계를 구한 결과, 각각 -23.7592, 12834.97819, -2138.126991로 계산되었다. 여기서 PORT6_CHN1에 대해서만 매우 큰 양수값으로 계산되었다. 그리고 이보다 작은 합계값을 가지는 윈도우는 300개 전부 해당하는 것으로 보아, PORT6_CHN1의 데이터는 대부분은 0으로부터 어느 정도 차이가 있는 양수값인 것으로 판단할 수 있다. 마찬가지로 PORT7_CHN1은 세 가지의 파라미터 중 가장 작은 음수값으로 계산되었고, 이보다 작은 합계값을 가지는 윈도우는 0개로 나타났다. 여기서 PORT7_CHN1의 데이터는 대부분 0으로부터 차이가 있는 음수값임을 알 수 있다. 해당 결과를 통해, 앞서 측정한 신뢰구간 95%와 90%에 해당하는 윈도우가 매우 적은 이유 중 하나를 근거로 삼을 수 있다.

D) Condition 4

- 각 윈도우 별 구간에 대한 합이 0보다 큰 윈도우 개수

```python
for i in range(3):
    if i is 0:
        print("PORT4_CHN1:")
    elif i is 1:
        print("PORT6_CHN1:")
    elif i is 2:
        print("PORT7_CHN1:")
        
    query2 = "SELECT COUNT(SUM) FROM statistic where sensor_ID='1' AND reg_ID=%s AND SUM > 0"
    cur.execute(query2, i+1)
    result = cur.fetchall()
    print(result[0])
```

=> 네 번째 조건으로는, 세 번째 조건과 마찬가지로 각 레지스터에 대해 전체 데이터의 합계보다 큰 합계를 가지고 있는 윈도우의 개수를 파악하는 과정이다. 신뢰구간 90%와 95%에 해당하는 윈도우가 매우 적은 이유에 대해 한 가지 더 근거를 들 수 있는 조건이라 할 수 있다. 먼저 PORT4_CHN1, PORT6_CHN1, PORT7_CHN1에 대하여 윈도우 구간에 대한 합이 0보다 큰 윈도수 개수는 각각 128, 196, 157으로 측정되었다. 이는 앞서 세 번째 조건에서 구한 각 파라미터에 대한 합계값(sum)과 비교해볼 수 있는데, 'PORT7_CHN1'에선 가장 큰 값이 나왔고 'PORT6_CHN1'에 대해선 가장 작은 값으로 계산된 결과가 있다. 우선 PORT6의 합계값에선 양수를 가지며 가장 큰 값을 가지는데, 윈도우 구간에 대한 합이 0보다 큰 세 번째 조건 결과에 맞게 윈도우 개수는 가장 큰 것으로 나타났다. 그리고 PORT7의 합계값은 음수값을 가지며 가장 작은 값을 나타내는데, 네 번째 조건의 결과와 비교해보면 합이 0보다 큰 윈도우 개수가 PORT4보다 큰 것으로 나타났다. 세 번째 조건으로 인하여, 네 번째 조건에선 가장 작은 결과값이 도출될 것이라 예상했었지만 결과는 예상과 다르게 나타났다. 이에 따라 PORT7에 대한 전체 데이터는 PORT4보다 훨씬 더 작은 음수의 데이터 값을 가지는 것을 알 수 있다. 따라서 위의 결과를 통해 각 파라미터에 대한 값 분포를 대략적으로 추측 가능하다.

E) Condition 5

- 각 레지스터의 전체 합계에 대해 10% 이내의 오차를 가지고 있는 윈도우 개수

```python
import pandas as pd

query = "SELECT SUM(PORT4_CHN1), SUM(PORT6_CHN1), SUM(PORT7_CHN1) FROM accelerator"
cur.execute(query)
SUM = cur.fetchall()
print(pd.DataFrame(SUM, columns= ['PORT4_sum', 'PORT6_sum', 'PORT7_sum']))

for i in SUM[0]:
    if i is SUM[0][0]:
        query2 = "SELECT COUNT(SUM) FROM statistic where sensor_ID='1' AND reg_ID='1' AND (SUM < %s*0.90 AND SUM > %s*1.1)"
        print("PORT4_CHN1:")
    elif i is SUM[0][1]:
        query2 = "SELECT COUNT(SUM) FROM statistic where sensor_ID='1' AND reg_ID='2' AND (SUM > %s*0.90 AND SUM < %s*1.1)"
        print("PORT6_CHN1:")
    elif i is SUM[0][2]:
        query2 = "SELECT COUNT(SUM) FROM statistic where sensor_ID='1' AND reg_ID='3' AND (SUM < %s*0.90 AND SUM > %s*1.1)"
        print("PORT7_CHN1:")

    cur.execute(query2, (i, i))
    result = cur.fetchall()
    print(result[0])
```

=> 다섯번째 조건은, 각 레지스터의 합계에 대하여 신뢰구간 90%에 속하는 윈도우 개수를 파악하는 과정이다. condition 1과 동일한 과정으로 구하였으며, 앞서 진행하였던 condition 3과 4에 대해 더 깊이 있게 파악하기 위해 진행하였다. 각 파라미터에 대해 결과를 측정하였을 때 'PORT4_CHN1' 만 17개의 윈도우가 신뢰구간에 속하고, 나머지 파라미터의 윈도우는 신뢰구간에 속하지 않음을 알 수 있다. 이는 앞 condition에 이어서 모든 파라미터의 데이터가 구간 별로 차이가 크다는 것을 뒷받침해주는 조건이라 할 수 있고, 이는 곧 각 파라미터의 데이터는 분산이 매우 크다는 것으로 정의할 수 있다.
